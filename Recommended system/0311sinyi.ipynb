```python
import pandas as pd
data = pd.read_csv("./sinyi_clean.csv")
```

## 抓出要下去算歐式距離的欄位

```python
data1 = data[["size", "totlefloor", "floor", "year", "room", "hall", "bath", "unitprice"]]
```

## 直接下去算歐式距離

```python
#計算歐式距離
from sklearn.metrics.pairwise import euclidean_distances

keys = []
values = []

for i in range(1, len(data1)):
    key = data["id"][i]
    house = data["houseLink"][i]
    value = euclidean_distances(data1.values[0].reshape(1,-1), data1.values[i].reshape(1,-1))[0][0]
    
    keys.append(key)
    values.append(value)
    
column = ['key','value']
df = pd.DataFrame(columns = column)

df['key'] = keys
df['value'] = values

df
```

```python
#由大排到小
df.sort_values("value", ascending = False, inplace = True) 
```

```python
## 看出差了很多
df
```

## 所以我決定先做標準差

```python
#做標準化
from sklearn import preprocessing
#建立MinMaxScaler物件
minmax = preprocessing.MinMaxScaler()
# 資料標準化
data_minmax = minmax.fit_transform(data1)
data_minmax_dataframe = pd.DataFrame(data_minmax, columns = ["size", "totlefloor", "floor", "year", "room", "hall", "bath", "unitprice"])
```

```python
data_minmax_dataframe
```

## 再去計算歐式距離

```python
#計算歐式距離
#計算第一個index跟其他人的距離

from sklearn.metrics.pairwise import euclidean_distances

idd = []
houseLink = []
pictureurl = []
values = [] 

for i in range(1, len(data_minmax_dataframe)):
    
    idd.append(data["id"][i])
    houseLink.append(data["houseLink"][i])
    pictureurl.append(data["pictureurl"][i])
    
    #0 input index
    value = euclidean_distances(data_minmax_dataframe.values[0].reshape(1,-1), data_minmax_dataframe.values[i].reshape(1,-1))[0][0]
    values.append(value)
    

column = ['id','houseLink', 'pictureurl', 'values']
df = pd.DataFrame(columns = column)

df['id'] = idd
df['houseLink'] = houseLink
df['pictureurl'] = pictureurl
df['value'] = values

```

```python
#取出前三名
id_output = df.nsmallest(3,'value')['id'].tolist()
houseLink_output = df.nsmallest(3,'value')['houseLink'].tolist()
pictureurl_output = df.nsmallest(3,'value')['pictureurl'].tolist()
value_output = df.nsmallest(3,'value')['value'].tolist()
```
