```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

data_path = 'all_new2-1.csv'
df_train = pd.read_csv(data_path)

df_train
```

```python
#處理'交易年月日'=>'年','月'
import pandas as pd
import numpy as np

df_train_1 = pd.DataFrame() #先創一個空的DataFrame
df_train_1['建築完成年'] = df_train['交易年月日'].apply(str).apply(lambda x: x[:-4])
df_train_1['建築完成年'].isnull().any()
df_train_1['建築完成年'][df_train_1['建築完成年']==''] = '0' #因為資料有兩筆空值!!!先取 0
df_train_1['建築完成年'] = df_train_1['建築完成年'].apply(int)
df_train_1['建築完成年']
```

```python
#處理'交易筆棟數'=>'交易筆棟數_土地','交易筆棟數_建物','交易筆棟數_車位'
import pandas as pd
import numpy as np
import re

df_train_1['交易筆棟數_土地'] = df_train['交易筆棟數'].apply(lambda x:int(re.findall(r'\d+',x)[0])) 
df_train_1['交易筆棟數_建物'] = df_train['交易筆棟數'].apply(lambda x:int(re.findall(r'\d+',x)[1]))
df_train_1['交易筆棟數_車位'] = df_train['交易筆棟數'].apply(lambda x:int(re.findall(r'\d+',x)[2]))

display(df_train_1)
```
